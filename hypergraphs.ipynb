```python
import hypernetx as hnx
import matplotlib.pyplot as plt

relations = {
    'FAMILY1': ('Adam', 'Amy', 'Alan'),
    'FAMILY2': ('Bob', 'Barbara', 'Billy'),
    'FAMILY3': ('Tyler', 'Terri', 'Teresa'),
    'COLLEGE': ('Amy', 'Barbara', 'Bob'),
    'SCHOOL': ('Alan', 'Billy', 'Teresa'),
    'WORK': ('Adam', 'Barbara', 'Tyler')
}    # <HYPEREDGE NAME>: (<LIST OF NODES CONNECTED BY THE HYPEREDGE)

H = hnx.Hypergraph(relations)
hnx.draw(H)
```
