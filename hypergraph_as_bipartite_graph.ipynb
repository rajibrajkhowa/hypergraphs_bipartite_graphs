```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms import bipartite


G = nx.Graph()
G.add_nodes_from(["FAMILY1","FAMILY2","FAMILY3","COLLEGE","SCHOOL","WORK"], bipartite=0)
G.add_nodes_from(["Adam","Amy","Alan","Bob","Barbara","Billy","Tyler","Terri","Teresa"], bipartite=1)
G.add_edges_from([("FAMILY1","Adam"),("FAMILY1","Amy"),("FAMILY1","Alan"),("FAMILY2","Bob"),("FAMILY2","Barbara"),("FAMILY2","Billy"), ("FAMILY3","Tyler"),("FAMILY3","Terri"),("FAMILY3","Teresa"),("COLLEGE","Amy"),("COLLEGE","Barbara"),("COLLEGE","Bob"),("SCHOOL","Alan"),("SCHOOL","Billy"),("SCHOOL","Teresa"),("WORK","Adam"),("WORK","Barbara"),("WORK","Tyler")])
X, Y = bipartite.sets(G)
pos = dict()
pos.update( (n, (1, i)) for i, n in enumerate(X) ) 
pos.update( (n, (2, i)) for i, n in enumerate(Y) )
nx.draw(G, pos=pos, with_labels=1)
plt.show()
```
